```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

pd.set_option("display.precision", 2)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
sns.set()

dados_vendas_censo = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/alura imersao 4/dados_vendas_censo.csv")
dados_vendas_censo = dados_vendas_censo[['Rua', 'Bairro', 'Cidade', 'Metragem', 'Quartos', 'Banheiros', 'Vagas',
       'Valor_anuncio', 'Valor_mm', 'Valor_m2', 'latitude', 'longitude', 'Situacao_setor', 'V001',
       'V002', 'V003', 'V004', 'V005', 'V006', 'V007', 'V008', 'V009', 'V010',
       'V011', 'V012']].dropna()
dados_vendas_censo
```

```python
plt.figure(figsize=(10, 8))
ax = sns.boxplot(data = dados_vendas_censo, y="Valor_anuncio")
plt.show()
```

```python
#dados_vendas_censo.query("Valor_anuncio > 30000000 | Valor_anuncio < 100000")
dados_vendas_censo.info()
```

```python
#dados_vendas_censo.drop(dados_vendas_censo.query("Valor_anuncio > 30000000 | Valor_anuncio < 100000").index, inplace = True)
#plt.figure(figsize=(10, 8))
#ax = sns.boxplot(data = dados_vendas_censo, y="Valor_anuncio")
#plt.show()

#i will be using the Interquartile Range method for dealing with outliers instead of choosing values by myself
df_IQR = dados_vendas_censo
for n in df_IQR.iloc[:,3:8]:
  nseries = df_IQR[[n]]
  nQ1 = nseries.quantile(0.25)
  nQ3 = nseries.quantile(0.75)
  nIQR = nQ3 - nQ1
  df_IQR = df_IQR[~((nseries < (nQ1 - 1.5 * nIQR)) | (nseries > (nQ3 + 1.5 * nIQR))).any(axis=1)]

f, axes = plt.subplots(nrows=2, ncols=5)
count_IQR = 0
for n in df_IQR.iloc[:,3:8]:
  dados_vendas_censo[[n]].plot.box(ax=axes[0, count_IQR], subplots = True)
  df_IQR[[n]].plot.box(ax=axes[1,count_IQR], subplots = True)
  count_IQR+=1

#top plots are from before applying IQR; bottom ones, after it: much cleaner and readable data
```
