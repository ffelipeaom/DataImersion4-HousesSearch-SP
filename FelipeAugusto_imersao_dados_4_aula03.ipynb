```python
ibge_dados = pd.read_csv(url_ibge, sep = ";", encoding="utf-8", decimal=',')
ibge_dados.dropna(axis=1)
ibge_dados.info()
```

```python
#1 link the data frames
ibge_dados.rename(columns = {"Nome_do_distrito":"Bairro"},inplace = True)
ibge_dados["Bairro"].unique()
```

```python
ibge_dados = ibge_dados.replace({"Bairro" : {'BEL�M' : 'BELÉM',
                      'BRASIL�NDIA' :	'BRASILÂNDIA',
                      'BR�S' : 'BRÁS',
                      'BUTANT�' :	'BUTANTÃ',
                      'CAP�O REDONDO' : 'CAPÃO REDONDO',
                      'CARR�O' : 'CARRÃO',
                      'CONSOLA��O' :	'CONSOLAÇÃO',
                      'FREGUESIA DO �' : 'FREGUESIA DO Ó',
                      'GRAJA�' : 'GRAJAÚ',
                      'JAGUAR�' : 'JAGUARÉ',
                      'JARAGU�' : 'JARAGUÁ',
                      'JARDIM S�O LU�S' : 'JARDIM SÃO LUÍS',
                      'JARDIM �NGELA' : 'JARDIM ÂNGELA',
                      'JA�AN�' : 'JAÇANÃ',
                      'JOS� BONIF�CIO' : 'JOSÉ BONIFÁCIO',
                      'LIM�O' : 'LIMÃO',
                      'REP�BLICA' : 'REPÚBLICA',
                      'SACOM�' : 'SACOMÃ',
                      'SANTA CEC�LIA' : 'SANTA CECÍLIA',
                      'SA�DE' : 'SAÚDE',
                      'S�' : 'SÉ',
                      'S�O DOMINGOS' : 'VILA SÃO DOMINGOS',
                      'S�O LUCAS' : 'SÃO LUCAS',
                      'S�O MATEUS' : 'SÃO MATEUS',
                      'S�O MIGUEL' : 'SÃO MIGUEL',
                      'S�O RAFAEL' : 'SÃO RAFAEL',
                      'TATUAP�' : 'TATUAPÉ',
                      'TREMEMB�' : 'TREMEMBÉ',
                      'VILA CURU��' : 'VILA CURUÇÁ',
                      'VILA JACU�' : 'VILA JACUÍ',
                      'VILA S�NIA' : 'VILA SÔNIA',
                      '�GUA RASA' : 'ÁGUA RASA'}})
```

```python
#dados_bairro["Bairro_upper"] = dados_bairro["Bairro"].str.upper()
dados_bairro["Bairro"] = dados_bairro['Bairro'].str.upper()
dados_bairro["Bairro"]
```

```python
ibge_dados['Bairro'] = ibge_dados['Bairro'].str.upper()
ibge_dados['Bairro']
```

```python
merge_dados = dados_bairro.merge(ibge_dados, on ='Bairro', how = 'left')
#merge_dados.sample(5)
merge_dados.info()
```

```python
#2 deal with outliers
#solved this one on the previous challenge by using Interquartile Range (IQR)
merged_IQR = merge_dados
for n in merge_dados.iloc[:,[1,2,3,4,7]]:
  nseries = merged_IQR[[n]]
  nQ1 = nseries.quantile(0.25)
  nQ3 = nseries.quantile(0.75)
  nIQR = nQ3 - nQ1
  merged_IQR = merged_IQR[~((nseries < (nQ1 - 1.5 * nIQR)) | (nseries > (nQ3 + 1.5 * nIQR))).any(axis=1)]
```

```python
#plotting the difference in Metragem (footage), before and after applying the IQR method
sns.set(rc = {'figure.figsize':(12,8)})
f, axes = plt.subplots(nrows=2, ncols=5)
count = 0
for n in merged_IQR.iloc[:,[1,2,3,4,7]]:
  merge_dados[[n]].plot.box(ax=axes[0, count], subplots = True)
  merged_IQR[[n]].plot.box(ax=axes[1,count], subplots = True)
  count+=1
#top plots are from before working the data; bottom ones after it
```

```python
#3 group into categories and analyze some relations
plot_data = merged_IQR.groupby(['Bairro']).mean().reset_index()

plt.figure(figsize=(18, 10))
for i, b in enumerate(plot_data["Bairro"]):
  plt.annotate(b, (plot_data["V003"][i], plot_data["Valor_m2_bairro"][i]), bbox=dict(boxstyle="round", fc=(1,1,1, 0.5), ec="b"), size=8)

ax1 = sns.scatterplot(data=plot_data, x="V003", y="Valor_m2_bairro", hue="Metragem", size="Metragem", palette = "YlOrBr_r", sizes=(150, 1500))
ax1.set_ylabel("Média do valor do m²")
ax1.set_xlabel("Média de moradores")
```
