```python
import folium
from folium import Choropleth, Marker
from folium.plugins import MarkerCluster

# creating a map
center = [-23.55, -46.63]
m_stamem = folium.Map(location=center, 
                 width=500,height=500,
                 tiles='Stamen Toner', zoom_start=12)
m_stamem
```

```python
#1 income distribution
city_name = "OSASCO"
city = setor_censo[setor_censo.NM_MUNICIP == city_name].reset_index(drop=True)
ax_city = city.plot(figsize=(8,12))
ax_city.set_title('Setores Censitários - Cidade de '+city_name);
```

```python
ibge_sp.Cod_setor = ibge_sp.Cod_setor.astype(str)
city_df = setor_censo_sp.merge(ibge_sp, left_on='CD_GEOCODI', right_on='Cod_setor', how='left')
#print(f'type: {type(sp_df)}')
city_df.info()
```

```python
#adding markers to centroids
city_df['centroid'] = city_df['geometry'].apply(lambda z: (z.centroid.y, z.centroid.x))
# Add points to the map
mc = MarkerCluster()
for idx, row in city_df.iterrows():
    mc.add_child(Marker(row['centroid'], popup=row['NM_BAIRRO']))

m_stamem.add_child(mc)
# Display the map
m_stamem
```

```python
#
city_plot = city_df[city_df.NM_MUNICIP == city_name].reset_index(drop=True)
ax2 = city_plot.plot(figsize=(20,10), column='V005', cmap='PuRd', edgecolor='gray', legend=True)
ax2.set_title('Classificação Coroplética da Renda nos Setores Censitários - '+city_name)
```

```python
#districts
district_name = "VILA MARIANA"
#district = setor_censo[setor_censo.NM_DISTRIT == district_name].reset_index(drop=True)
#ax_district = district.plot(figsize=(8,12))
#ax_district.set_title('Setores Censitários - Cidade de '+district_name);

district_plot = city_df[city_df.NM_DISTRIT == district_name].reset_index(drop=True)
ax2 = city_plot.plot(figsize=(20,10), column='V005', cmap='PuRd', edgecolor='gray', legend=True)
ax2.set_title('Classificação Coroplética da Renda nos Setores Censitários - '+district_name)
```

```python
# Create the map
m_stamen_2 = folium.Map(location=center, 
                 tiles='Stamen Toner', zoom_start=12)

# Add a choropleth map to the base map
Choropleth(geo_data=city.__geo_interface__, 
           data=city_df, 
           columns=['CD_GEOCODI', 'V005'],
           key_on="feature.properties.CD_GEOCODI", 
           fill_color='BuGn', 
           fill_opacity=0.5,
           legend_name='Renda Média'
          ).add_to(m_stamen_2)
# Display the map
m_stamen_2
```
